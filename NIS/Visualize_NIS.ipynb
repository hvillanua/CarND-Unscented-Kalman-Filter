```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
laser_path = './NISvals_laser.txt'
radar_path = './NISvals_radar.txt'

laser_f = open(laser_path)
radar_f = open(radar_path)
laser_x = list(map(float, laser_f.read().split()))
radar_x = list(map(float, radar_f.read().split()))
laser_y = list(range(len(laser_x)))
radar_y = list(range(len(radar_x)))

laser_limit = 6
radar_limit = 7.8
```

```python
plt.plot(laser_y, laser_x)
plt.axhline(y=laser_limit, xmin=0, xmax=len(laser_x), c='red')
plt.title('Laser NIS')
plt.savefig('laser_NIS.jpg')
plt.show()

plt.plot(radar_y, radar_x)
plt.axhline(y=radar_limit, xmin=0, xmax=len(radar_x), c='red')
plt.title('Radar NIS')
plt.savefig('radar_NIS.jpg')
plt.show()
```
